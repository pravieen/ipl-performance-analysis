```python
bat_first = ipl[ipl["win_by_runs"] != 0]
```

```python
bat_first.head()
```

```python
plt.figure(figsize=(7,7))
plt.hist(bat_first['win_by_runs'])
plt.xlabel("Runs")
plt.ylabel("matches")
plt.title("Runs distribution")

plt.show()
```

```python
bat_first['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(
    list(bat_first['winner'].value_counts().keys()[0:3]),
    list(bat_first['winner'].value_counts()[0:3]),
    color=['blue','yellow','orange']
    )

plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(bat_first['winner'].value_counts()),labels=list(bat_first['winner'].value_counts().keys()),autopct="%.1f%%")
plt.show()
```

```python
bat_second=ipl[ipl["win_by_wickets"]!=0]
```

```python
plt.hist(bat_second['win_by_wickets'],bins=30)
plt.xlabel("Wickets")
plt.ylabel("matches")
plt.title("Wickets Distribution")
plt.show()
```

```python
bat_second['winner'].value_counts()
```

```python
plt.bar(
    list(bat_second['winner'].value_counts()[0:3].keys()),
    list(bat_second['winner'].value_counts()[0:3]),
    color=['purple','blue','red']
    )
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(
    list(bat_second['winner'].value_counts()),
    labels=list(bat_second['winner'].value_counts().keys()),
    autopct="%.1f%%"
    )
plt.show()
```
